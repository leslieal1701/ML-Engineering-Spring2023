# Original database

```python
maindat = pd.read_csv("NonquantumONLY.csv")
maindat.head()
```

```python
df1 = pd.read_csv("batch1_0_27.csv")
df2 = pd.read_csv("batch2_28_39.csv")
df3 = pd.read_csv("batch3_40_65.csv")
df4 = pd.read_csv("batch4_66_72.csv")
df5 = pd.read_csv("batch5_73_90.csv")
df6 = pd.read_csv("batch6_91_117.csv")
df7 = pd.read_csv("batch7_118_132.csv")
df8 = pd.read_csv("batch8_133_to_140.csv")
df9 = pd.read_csv("batch8p5_141_149.csv")
df10 = pd.read_csv("batch9_150_167.csv")
df11 = pd.read_csv("batch10_168_all.csv")


combined = pd.concat([df1,df2,df3,df4,df5,df6,df7,df8,df9,df10,df11], axis=0,ignore_index=True)
combined.drop(['Fuel'],axis='columns',inplace=True)
```
